```python
# S&P 500
url_sp500 = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
sp500 = pd.read_html(url_sp500)[0]  
sp500["Index"] = "S&P 500"

# NASDAQ-100
url_nasdaq = "https://en.wikipedia.org/wiki/NASDAQ-100"
nasdaq100 = pd.read_html(url_nasdaq)[4]  
nasdaq100 = nasdaq100.rename(columns={"Ticker": "Symbol"})
nasdaq100["Index"] = "NASDAQ-100"

# Dow Jones
url_dow = "https://en.wikipedia.org/wiki/Dow_Jones_Industrial_Average"
dow = pd.read_html(url_dow)[2]
dow["Index"] = "Dow Jones"
```

```python
df_all = pd.concat([
    sp500[["Symbol"]],
    nasdaq100[["Symbol"]],
    dow[["Symbol"]],
])
df_all = df_all.drop_duplicates(subset="Symbol")
tickers = df_all["Symbol"].tolist()
len(tickers)
```

```python
tickers
```

```python
tickers[:12]
```

```python
sublistas = [tickers[i:i + 10] for i in range(0, len(tickers), 10)]
print(f"Nmero de sublistas: {len(sublistas)}")
```

```python
for i, bloque in enumerate(sublistas):
    print(i, bloque)
```

```python
atributos = [
    "marketCap",
    "sector",
    "trailingPE",
    "forwardPE",
    "priceToBook",
    "dividendYield",
    "beta",
    "profitMargins",
    "returnOnEquity",
    "debtToEquity",
    "currentRatio",
    "earningsGrowth",
    "trailingPegRatio",
    "returnOnAssets",
    "epsForward",
]
```

```python
dfs = []  # Ac guardamos todos los dataframes

for i, bloque in enumerate(sublistas):
    datos = []  # Se crea una nueva lista vaca para cada bloque
    for ticker in bloque:
        info = yf.Ticker(ticker).info
        data = {atributo: info.get(atributo, None) for atributo in atributos}
        data["Ticker"] = ticker
        datos.append(data)
    
    df_bloque = pd.DataFrame(datos).set_index("Ticker")
    dfs.append(df_bloque)  # Guardamos el DataFrame para este bloque

# Creamos un nico dataframe con todos los datos
df_final = pd.concat(dfs)
```
